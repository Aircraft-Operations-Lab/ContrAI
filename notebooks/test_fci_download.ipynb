```python
import eumdac
from datetime import datetime
from contrai.data.mtg_fci import download_fci_l1c_products
import os
# Path to credentials file
cred_path = os.path.expanduser("~/.eumdac/credentials")
# Read key/secret from file
credentials = {}
with open(cred_path, "r") as f:
    for line in f:
        consumer_key, consumer_secret = line.strip().split(",")

# Create AccessToken
token = eumdac.AccessToken((consumer_key, consumer_secret))
datastore = eumdac.DataStore(token)

# Example usage
files = download_fci_l1c_products(
    datastore=datastore,
    start=datetime(2025, 1, 5, 10, 0),
    end=datetime(2025, 1, 5, 11, 0),
    out_dir="mtg_fci_l1c",
    limit=4,
)
```
